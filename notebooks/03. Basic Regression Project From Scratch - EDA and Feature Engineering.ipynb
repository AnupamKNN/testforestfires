### Algerian Forest Fires Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header = 1)
```

```python
dataset.head()
```

```python
dataset.info()
```

### Data Cleaning

```python
## Missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region. From 122h index, we can make a new column based on the region

"Bejaia Region Dataset"

"Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"] = 1
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## Fix column names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

### Change the required columns as integer data type

```python
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] = df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

### Change the other columns to float data type

```python
objects = [features for features in df.columns if df[features].dtypes == 'O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.head()
```

```python
## Let us save the cleaned data
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index = False)
```

### Exploratory Data Analysis

```python
## drop day, month and year
df_copy = df.drop(columns = ['day', 'month', 'year'], axis = 1)
```

```python
df_copy.head()
```

```python
## Categories in Classes
df_copy['Classes'].value_counts()
```

```python
## Encoding of the categories in Classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
df.copy().head()
```

```python
df_copy.tail()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Plot density plot for all the features
plt.style.use('fivethirtyeight')
df_copy.hist(bins = 50, figsize=(20, 15))
plt.show()
```

```python
## Percentage for Pie Chart
percentage = df_copy['Classes'].value_counts(normalize = True) * 100
```

```python
## plotting pie chart
classlabels = ['Fire', 'Not Fire']
plt.figure(figsize = (12, 7))
plt.pie(percentage, labels= classlabels, autopct= '%1.1f%%')
plt.title('Pie Chart for Classes')
plt.show()
```
